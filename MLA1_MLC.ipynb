# 6. Display Column Names

```python
print("\nList of all columns in the dataset:")
print(df.columns)
```

# 7. Rename Column Names

```python
df.columns = [
    'date_time', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visability', 'pressure', 'weather'
]
```

# 8. Display Dataset Statistics

```python
print("\nDataset statistics:")
print(df.describe())
```

# 9. Convert Celsius to Fahrenheit and Add a New Column

```python
df['temp_f'] = (df['temp_c'] * 9/5) + 32
```

# 10. Rearrange columns so temp_f appears before temp_c

```python
column_order = ['date_time', 'temp_f', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visability', 'pressure', 'weather']
df = df[column_order]
```

# 11. Display Data Types

```python
print("\nData types of all columns:")
print(df.dtypes)
```

# 12. Convert date_time to a proper datatime format

```python
df['date_time'] = pd.to_datetime(df['date_time'])
```

```python
print("\Data type after the conversion:")
print(df.dtypes)
```

```python
df['month'] = df['date_time'].dt.month
```

# 13. Calculate the Average Temperature (F) Per Month

```python
monthly_avg = df.groupby('month')['temp_f'].mean()
```

```python
print("\Monthly average tempature (F):")
print(monthly_avg)
```

# 14. Generate a Scattterplot Comparing Temperature vs Relative Humidity

```python
plt.figure(figsize=(20, 10))
sns.scatterplot (x=df['temp_f'], y=df['relative_humidity'])
plt.xlabel('Temperature (F)')
plt.ylabel('Relative Humidity (%)')
plt.title('Temperature vs. Relative Humidity')
plt.show()
```

# 15. Generate a Line Chart Showing the Average Temperature per Month

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x=monthly_avg.index, y=monthly_avg.values, marker='o')
plt.xlabel('Month')
plt.ylabel('Average Temperature (F)')
plt.title('Average Monthly Temperature')
plt.xticks(range(1, 13))
plt.grid()
plt.show
```

# 16. Generate a Box Plot of Temperature

```python
plt.figure(figsize=(20, 10))

sns.boxplot(data=df[['temp_f', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visability', 'pressure']])

plt.title('Box Plot of Weather Variables')
plt.ylabel('Value')
plt.xlabel('Weather Variables')

plt.show
```
